## D3. Additional queries for Read me.

```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")
Base = automap_base()
Base.prepare(engine, reflect=True)
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
session = Session(engine)
```

```python
sept_precip = session.query(Measurement.date, Measurement.prcp).filter(func.extract('month', Measurement.date) == '9').all()
sept_df = pd.DataFrame(sept_precip, columns=['Date','Rain'])
sept_df.describe()
```

```python
sept_df.plot(figsize=(10,5))
plt.xticks(rotation=90)
plt.show()
```

```python
oct_precip = session.query(Measurement.date, Measurement.prcp).filter(func.extract('month', Measurement.date) == '10').all()
oct_df = pd.DataFrame(oct_precip, columns=['Date','Rain'])
oct_df.describe()
```

```python
oct_df.plot(figsize=(10,5))
plt.xticks(rotation=90)
plt.show()
```
